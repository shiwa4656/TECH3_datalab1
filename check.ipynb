```python
import matplotlib.pyplot as plt

# Output levels and average costs
Q_curve = [500, 1000, 2000, 3000, 4000, 5000, 6000]
AC_curve = [55, 40, 30, 24, 20, 18, 17]

plt.figure()
plt.plot(Q_curve, AC_curve)
plt.scatter([1000, 5000], [40, 18])

plt.xlabel("Output (Q)")
plt.ylabel("Average Cost (AC)")
plt.title("Average Cost Curve (Increasing Returns to Scale)")
plt.xlim(0, 6500)
plt.ylim(0, 60)

plt.show()
```
